```python
## import modules

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

## load dataset

demo_data = pd.read_csv("datasets/DEMO_data.csv") #demographic data
```

## Task 1: Create a box plot of the data in the *RIDAGEYR* column

```python
## pull out age values from dataframe and store in a list variable
value_list = demo_data["RIDAGEYR"]
```

```python
## copy and paste the box plot code here
## replace value_list with your list of age values
## give the plot an appropriate plot title and axes labels

fig, axs = plt.subplots(figsize=(12,8))
axs.boxplot(value_list)
plt.title("Plot Title", fontsize=20)
axs.set_xticklabels(["Box Plot Label"])
axs.set_ylabel("Label for the value represented in plot", fontsize=18)
axs.tick_params(labelsize=15)
plt.show()
```

## Task 2: Split data into race groups

```python
## create separate dataframes for each race groups
ma_df = demo_data[demo_data["RIDRETH3"] == 1][["RIDRETH3", "RIDAGEYR"]]
h_df = demo_data[demo_data["RIDRETH3"] == 2][["RIDRETH3", "RIDAGEYR"]]
w_df = demo_data[demo_data["RIDRETH3"] == 3][["RIDRETH3", "RIDAGEYR"]]
b_df = demo_data[demo_data["RIDRETH3"] == 4][["RIDRETH3", "RIDAGEYR"]]
a_df = demo_data[demo_data["RIDRETH3"] == 6][["RIDRETH3", "RIDAGEYR"]]
o_df = demo_data[demo_data["RIDRETH3"] == 7][["RIDRETH3", "RIDAGEYR"]]

## display the first 2 entries of each dataframe
print("Mexican American:")
display(ma_df.head(2))
print("Other Hispanic:")
display(h_df.head(2))
print("Non-Hispanic White:")
display(w_df.head(2))
print("Non-Hispanic Black:")
display(b_df.head(2))
print("Non-Hispanic Asian:")
display(a_df.head(2))
print("Other Race - Including Multi-Racial:")
display(o_df.head(2))
```

## Task 3: Create a comparative box plot

```python
## pull out age values for each race from the dataframe for that race
## and store this list in a variable

ma_age_values = ma_df['RIDAGEYR']
h_age_values = h_df['RIDAGEYR']
w_age_values = w_df['RIDAGEYR']
b_age_values = b_df['RIDAGEYR']
a_age_values = a_df['RIDAGEYR']
o_age_values = o_df['RIDAGEYR']
```
