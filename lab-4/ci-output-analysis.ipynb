##  Task 2

````python
# calculate the sample proportion of households with access to the internet

n = comm_data_sample.shape[0] #sample size

count_access = 0 #counter to keep track of no. households with internet access

for i in comm_data_sample.index:
    
    if comm_data_sample.loc[i, "ACCESS"] == 1 or comm_data_sample.loc[i, "ACCESS"] == 2:
        #this household has internet access
        #increment the counter
        count_access = count_access + 1

# calculate sample proportion
# p_hat =(no. households with internet access/total no. households)
p_hat = count_access/n

print("Sample proportion of households with access to the internet:", p_hat)
````

````python
# generate 95% confidence interval for population proportion

low_ci_bound = p_hat - 1.96 * math.sqrt(p_hat * (1-p_hat) /n)
high_ci_bound = p_hat + 1.96 * math.sqrt(p_hat * (1-p_hat) /n)

prop_ci = [low_ci_bound, high_ci_bound]
print("95% confidence interval for population proportion:", prop_ci)
````

## Task 3

````python
# generate 95% confidence interval for population mean

# n is the sample size

all_num_people_values = comm_data_sample["NP"]
num_people_mean = np.mean(all_num_people_values)
num_people_stdev = np.std(all_num_people_values)

low_ci_bound = num_people_mean - 1.96 * num_people_stdev / math.sqrt(n)
high_ci_bound = num_people_mean + 1.96 * num_people_stdev / math.sqrt(n)

mean_ci = [low_ci_bound, high_ci_bound]
print("95% confidence interval for population mean:", mean_ci)
````

## Task 4

````python
# create a multigenerational household sample and a non-multigenerational household sample

multi_gen = comm_data[comm_data["MULTG"] == 2]
non_multi_gen = comm_data[comm_data["MULTG"] == 1]
print("Multigenerational household sample:")
display(multi_gen.head())
print("Non-multigenerational household sample:")
display(non_multi_gen.head())
````

````python
# calculate the mean and stdev for each sample

multi_gen_np_vals = multi_gen["NP"]
multi_gen_mean = np.mean(multi_gen_np_vals)
multi_gen_stdev = np.std(multi_gen_np_vals)

non_multi_gen_np_vals = non_multi_gen["NP"]
non_multi_gen_mean = np.mean(non_multi_gen_np_vals)
non_multi_gen_stdev =  np.std(non_multi_gen_np_vals)

print("Mean number of people in multigenerational households:", multi_gen_mean)
print("Standard deviation for number of people in multigenerational households:", multi_gen_stdev)

print("Mean number of people in non-multigenerational households:", non_multi_gen_mean)
print("Standard deviation for number of people in non-multigenerational households:", non_multi_gen_stdev)
````

````python
# generate 95% confidence interval for population mean difference

n1 = multi_gen.shape[0]
n2 = non_multi_gen.shape[0]

low_ci_bound = (multi_gen_mean - non_multi_gen_mean) - 1.96*(
    math.sqrt((multi_gen_stdev**2)/n1 + (non_multi_gen_stdev**2)/n2))
high_ci_bound = (multi_gen_mean - non_multi_gen_mean) + 1.96*(
    math.sqrt((multi_gen_stdev**2)/n1 + (non_multi_gen_stdev**2)/n2))

mean_diff_ci = [low_ci_bound, high_ci_bound]
print("95% confidence interval for population mean difference:", mean_diff_ci)
````
